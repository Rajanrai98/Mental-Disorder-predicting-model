```python
x = data.drop(["Expert Diagnose"] , axis = 1)
y = data["Expert Diagnose"]
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 42)
```

```python
x_train.shape , x_test.shape
```

```python
model = DecisionTreeClassifier(criterion = "entropy" , max_depth = 3)
```

```python
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
pd.DataFrame({"y_test":y_test , "y_pred":y_pred})
```

```python
accuracy_tree = accuracy_score(y_test , y_pred)
print(accuracy_tree)
```

```python
classification_result = classification_report(y_test , y_pred)
print(classification_result)
```

```python
model_random = RandomForestClassifier(n_estimators = 100 , max_features = 5)
model_random.fit(x_train , y_train)
```

```python
y_random_pred = model_random.predict(x_test)
y_random_pred
                                     
```

```python
pd.DataFrame({"y_test":y_test , "y_random_pred":y_random_pred})
```

```python
accuracy_result = accuracy_score(y_test , y_random_pred)
```

```python
accuracy_result
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model_bagging = BaggingClassifier(model_random ,n_estimators = 10, random_state = 42 )
```

```python
model_bagging.fit(x_train , y_train)
```

```python
y_pred_bagging = model_bagging.predict(x_test)
```

```python
y_pred_bagging
```
